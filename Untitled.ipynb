```python
stlr = ascii.read("data/q1_q17_dr25_stellar.txt").to_pandas()
```

```python
cdpp_cols = ["rrmscdpp01p5", "rrmscdpp02p0", "rrmscdpp02p5", "rrmscdpp03p0",
             "rrmscdpp03p5", "rrmscdpp04p5", "rrmscdpp05p0", "rrmscdpp06p0",
             "rrmscdpp07p5", "rrmscdpp09p0", "rrmscdpp10p5", "rrmscdpp12p0",
             "rrmscdpp12p5", "rrmscdpp15p0"]
cdpp_vals = np.array([1.5, 2.0, 2.5, 3.0, 3.5, 4.5, 5.0, 6.0, 7.5, 9.0, 10.5, 12.0, 12.5, 15.0])

pars = stlr[cdpp_cols + ["kepid", "limbdark_coeff1", "limbdark_coeff2", "limbdark_coeff3", "limbdark_coeff4", "dataspan", "dutycycle"]]
```

```python
df = ascii.read("data/kplr_dr25_inj1_plti.txt").to_pandas()
df = df.append(ascii.read("data/kplr_dr25_inj2_plti.txt").to_pandas())
df = df.append(ascii.read("data/kplr_dr25_inj3_plti.txt").to_pandas())

m = np.isfinite(df.Measured_MES)
df = df[m]
```

```python
df["kepid"] = df.KIC_ID.astype(int)
```

```python
joined = pd.merge(df, pars, on="kepid")
joined = joined.iloc[:1000]
c2 = np.array(joined.limbdark_coeff2)
c4 = np.array(joined.limbdark_coeff4)
g1 = 0.1  # c2 + 2*c4
g2 = 0.2  # -c4
ror = np.array(joined.i_ror)
b = np.array(joined.i_b)
joined["depth"] = (1 - quad.quad(g1, c2, ror, b)) * 1e6
```

```python
plt.loglog(joined.depth, joined.i_depth, ".")
plt.plot(joined.depth, joined.depth)
```

```python
from tqdm import tqdm
joined["mes_model"] = np.nan + np.zeros(len(joined))
for i, star in tqdm(joined.iterrows(), total=len(joined)):
    kicid = star.kepid
    fn = "pdm/kplr{0:09d}_dr25_onesigdepth.fits".format(kicid)
    if not os.path.exists(fn):
        continue
    duration = star.i_dur
    depth = star.depth
    ind = int(np.digitize(duration, cdpp_vals))
    with fits.open(fn) as f:
        if ind >= len(f)-1:
            x = f[-1].data["PERIOD"]
            y = f[-1].data["ONESIGDEP"]
            osd = np.interp(star.i_period, x, y)
        elif ind == 0:
            x = f[1].data["PERIOD"]
            y = f[1].data["ONESIGDEP"]
            osd = np.interp(star.i_period, x, y)
        else:
            if f[ind].data is None:
                continue
            x = f[ind].data["PERIOD"]
            y = f[ind].data["ONESIGDEP"]
            osd_m = np.interp(star.i_period, x, y)
            x = f[ind+1].data["PERIOD"]
            y = f[ind+1].data["ONESIGDEP"]
            osd_p = np.interp(star.i_period, x, y)
            osd = osd_m + (osd_p - osd_m) * (duration - cdpp_vals[ind-1]) / (cdpp_vals[ind] - cdpp_vals[ind-1])
    
    joined.loc[i, "mes_model"] = depth / osd
    
    if depth / osd < 0:
        print(depth, osd)
        assert 0
```

```python
X = np.array(df[["i_ror", "i_period", "i_b"]])
m = X[:, 2] > 0.0
X[~m, 2] = 0.5 * X[m, 2].min()
X = np.log(X)
mu = np.mean(X, axis=0)
X -= mu
y = np.log(np.array(df.Measured_MES))
```

```python
order = 2
As = [np.concatenate((np.ones((len(X), 1)), X), axis=1)]
for order in range(2, 3):
    As.append(np.product(list(
        combinations_with_replacement(X.T, order)), axis=1).T)
A = np.concatenate(As, axis=1)
X.shape, A.shape
```

```python
np.random.seed(42)
train = np.random.rand(len(X)) < 0.8
w = np.linalg.solve(np.dot(A[train].T, A[train]), np.dot(A[train].T, y[train]))
y_train = np.dot(A[train], w)
y_test = np.dot(A[~train], w)
```

```python
plt.plot(y[train], y[train] - y_train, ".")
plt.plot(y[~train], y[~train] - y_test, ".")
```
